```python
def euclidean_algorithm(a, b):
    amod = a
    bmod = b
    # Перевірка на нульові значення
    if amod == 0 or bmod == 0:
        return "Один або обидва з чисел дорівнює нулю. Неможливо знайти найбільший спільний дільник."

    # Функція для знаходження найбільшого спільного дільника
    while bmod != 0:
        amod, bmod = bmod, amod % bmod

    return f"Найбільший спільний дільник {a} і {b} - це {amod + bmod}"
```
